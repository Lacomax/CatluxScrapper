```python
from urllib.parse import urljoin
import requests
from bs4 import BeautifulSoup
from pathlib import Path

CERT_PATH = r'C:\Users\lacomax\OneDrive - Telefonica\Desktop\Catlux\catlux.de.crt'
SAVE_BASE_PATH = r'C:\Users\lacomax\OneDrive - Telefonica\Desktop\Catlux'
LOGIN_URL = "https://www.catlux.de/login"
username = "xavierlacoma@gmail.com"
password= "Familia2200"

def download_filtered_pdfs(base_url, page_param, max_pages):
    class_folder, subject_folder = base_url.split('/')[-3:-1]
    full_save_path = Path(SAVE_BASE_PATH, class_folder, subject_folder)
    full_save_path.mkdir(parents=True, exist_ok=True)
    
    with requests.Session() as session:
        login_page_req = session.get(LOGIN_URL, verify=CERT_PATH)
        soup_login = BeautifulSoup(login_page_req.content, 'html.parser')
        request_token = soup_login.find('input', {'name': 'REQUEST_TOKEN'})['value']
        
        payload = {
            'FORM_SUBMIT': 'tl_login',
            'REQUEST_TOKEN': request_token,
            'username': username,
            'password': password
        }
        
        login_req = session.post(LOGIN_URL, data=payload, verify=CERT_PATH)
        
        if login_req.status_code != 200:
            print("Login fehlgeschlagen.")
            return

        for page_num in range(1, max_pages + 1):
            url = f"{base_url}?{page_param}={page_num}"
            print(url)
            response = session.get(url, verify=CERT_PATH)
            soup = BeautifulSoup(response.content, "html.parser")
            pdf_urls = [link['href'] for link in soup.select('a[href*="pdf"]')]
            
            for url in pdf_urls:
                full_url = urljoin("https://www.catlux.de/", url)
                pdf_name = Path(url).name.replace("?dl=pdf", "") + '.pdf'
                pdf_save_path = full_save_path / pdf_name
                
                if not pdf_save_path.exists():
                    r = session.get(full_url, verify=CERT_PATH)
                    with open(pdf_save_path, 'wb') as f:
                        f.write(r.content)
                    print(f"{pdf_name} heruntergeladen.")
                else:
                    print(f"{pdf_name} existiert bereits.")

if __name__ == '__main__':
    base_url = "https://www.catlux.de/proben/gymnasium/klasse-5/deutsch/"
    download_filtered_pdfs(base_url, "p", 10)
    print("Script ausgeführt, keine Fehler gefunden.")
```
