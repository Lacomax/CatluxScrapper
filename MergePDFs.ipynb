```python
import os
from reportlab.lib.pagesizes import letter
from reportlab.pdfgen import canvas
from PyPDF2 import PdfFileReader, PdfFileWriter
from io import BytesIO

def add_blank_page_if_needed(pdf_writer):
    if pdf_writer.getNumPages() % 2 != 0:
        packet = BytesIO()
        blank_canvas = canvas.Canvas(packet, pagesize=letter)
        blank_canvas.drawString(100, 100, "")
        blank_canvas.save()
        packet.seek(0)
        blank_page = PdfFileReader(packet).getPage(0)
        pdf_writer.addPage(blank_page)

def merge_pdfs(folder_path, output_filename):
    pdf_writer = PdfFileWriter()
    output_file_name_only = os.path.basename(output_filename)
    
    for pdf_file in os.listdir(folder_path):
        if pdf_file == output_file_name_only:
            print(output_file_name_only, " already existed and was ignored for the merge")
            continue
            
        pdf_reader = PdfFileReader(os.path.join(folder_path, pdf_file))
        
        for page_num in range(pdf_reader.getNumPages()):
            page = pdf_reader.getPage(page_num)
            pdf_writer.addPage(page)
            
        if pdf_reader.getNumPages() % 2 != 0:
            add_blank_page_if_needed(pdf_writer)
            
    with open(output_filename, 'wb') as out_pdf:
        pdf_writer.write(out_pdf)
    print(f"{output_filename} erstellt.")



    
folder_path = r'C:\Users\lacomax\Downloads\print 20230926'
output_filename = r'C:\Users\lacomax\Downloads\print 20230926\_merged_file.pdf'
merge_pdfs(folder_path, output_filename)
```
